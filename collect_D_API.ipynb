```python
import requests
full_url="https://reqres.in/api/users"
response=requests.get(full_url)
response.status_code
data=response.json()
type(data)
data_users=data["data"]
```

```python
import csv
```

```python
with open ("users.csv","w",newline="") as file:
    writer=csv.DictWriter(file,["id","email","first_name","last_name","avatar"])
    writer.writeheader()
    for d in data_users:
        id=d["id"]
        email=d["email"]
        first_name=d["first_name"]
        last_name=d["last_name"]
        avatar=d["avatar"]
        writer.writerow({"id":id,"email":email,"first_name":first_name,"last_name":last_name,"avatar":avatar})
```

```python
import requests
full_url="http://universities.hipolabs.com/search"
response=requests.get(full_url)
response.status_code
data=response.json()
data
```

```python
import csv
```

```python
with open ("university_search.csv","w",newline="") as file:
    writer=csv.DictWriter(file,["country", "name","web_pages"])
    writer.writeheader()
    for d in data:
        country=d.get("country")
        name=d.get("name")
        web_pages=d.get("web_pages")
        writer.writerow({"country":country,"name":name,"web_pages":web_pages})
```
